```python
import pandas as pd, numpy as np
import variables_nombres as vn
```

```python
# Cargar datos
path = r'..\..\output\data_preprocess\dfs_0_i_mca.csv'
data = pd.read_csv( path )
```

```python
a = data.describe(include='all')
a = a.T
a
```

```python
a[ 'missing_values' ] = 652 - a[ 'count' ]
```

```python
a
```

```python
a[ 'missing_values_percentage' ] = a[ 'missing_values' ]/652
a = a.sort_values( by = "missing_values_percentage",
               ascending = False )
```

```python
fuente = []
for i, row in a.iterrows():
    if i in vn.siaf_variables: fuente.append( "SIAF" )
    if i in vn.renamu_variables: fuente.append( "Renamu" )
    if i not in vn.siaf_variables and i not in vn.renamu_variables: fuente.append( "Other" )
```

```python
a['fuente'] = fuente
a
```

```python
a.to_excel( 'table.xlsx' )
```
